```python
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
import matplotlib.pyplot as plt
```

```python
rawImage = misc.face(gray=True)
rawImage
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(rawImage, cmap=plt.cm.gray)
plt.show()
```

```python
from sklearn.cluster import KMeans
rawImageRe = rawImage.reshape(-1,1)
kmeans=KMeans(n_clusters=5,random_state=30)
kmeans.fit(rawImageRe)
```

```python
labels = kmeans.labels_
compress = np.asarray(labels)
compress.shape = rawImage.shape
```

```python
plt.figure(2)
plt.imshow(compress,cmap=plt.cm.gray)
```
